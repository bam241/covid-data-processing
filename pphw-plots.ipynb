```python
import ipywidgets as widgets
from ipywidgets import interact_manual
from IPython.display import display
import numpy as np
import pandas as pd
```

```python
#%matplotlib widget
#import ipympl
import matplotlib.pyplot as plt

plt.style.use('ggplot')

import plots

def generate(countries):
    try:
        plt.close('all')
        plots.generate_registered_plots([s.strip() for s in countries.split(',')])
    except Exception as e:
        print(e)
        
cases, _, _ = plots.parse_data()
print(f"Following countries are available {cases.index.values}")
default_countries = "Italy, US, Germany, Canada"
```

```python
interact_manual(generate, countries=default_countries)
```
